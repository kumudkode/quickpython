Conditional statements in Python are used to execute different blocks of code based on whether a condition is true or false. The primary conditional statements are `if`, `elif` (else if), and `else`.

**Syntax:**

```python
if condition:
    # Code to execute if the condition is true
elif another_condition:
    # Code to execute if the another_condition is true
else:
    # Code to execute if none of the above conditions are true
```

**Example:**

```python
x = 10
if x > 0:
    print("x is positive")
elif x < 0:
    print("x is negative")
else:
    print("x is zero")
```
```

**Advanced Conditional Statements**

1.  **Nested Conditionals:**
    *   You can nest `if` statements inside other `if` statements to create more complex decision-making processes.

    ```python
    x = 10
    y = 5
    if x > 0:
        if y > 0:
            print("Both x and y are positive")
        else:
            print("x is positive but y is not")
    else:
        print("x is not positive")
    ```

2.  **Chained Conditionals with Logical Operators:**

    *   You can combine multiple conditions using logical operators like `and`, `or`, and `not`.

    ```python
    age = 25
    if age > 18 and age < 60:
        print("You are eligible to work")
    ```

3.  **Conditional Expressions (Ternary Operator):**

    *   A more concise way to write simple `if-else` statements in a single line.

    ```python
    x = 10
    parity = "even" if x % 2 == 0 else "odd"
    print(f"x is {parity}")
    ```

4.  **`in` operator:**

    *   Check if a value exists in a sequence (e.g., list, tuple, string).

    ```python
    colors = ["red", "green", "blue"]
    if "red" in colors:
        print("Red is available")
    ```

5.  **Complex Conditional Logic:**

    *   Combining multiple techniques for advanced control flow.

    ```python
    # Example: Checking if a number is within a specific range and also even
    num = 24
    lower_bound = 20
    upper_bound = 30

    if lower_bound <= num <= upper_bound and num % 2 == 0:
        print(f"{num} is within the range {lower_bound}-{upper_bound} and is even.")
    else:
        print(f"{num} does not meet the criteria.")
    ```
